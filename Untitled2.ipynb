```python
df.head()
# df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.set_index('date')
df.drop(['date'], axis=1, inplace=True)
df.head()
```

```python
import tensorflow as tf
WINDOW_SIZE = 3
def batch_dataset(dataset):
    dataset_batched = dataset.batch(WINDOW_SIZE, drop_remainder=True)
    return dataset_batched
ds_data = tf.data.Dataset.from_tensor_slices(tf.constant(df.values, dtype=tf.float32)) \
.window(WINDOW_SIZE, shift=1).flat_map(batch_dataset)
ds_label = tf.data.Dataset.from_tensor_slices(tf.constant(df.values[WINDOW_SIZE:], dtype=tf.float32))
ds_train = tf.data.Dataset.zip((ds_data, ds_label)).batch(169).cache()
```

```python
print(len(list(ds_data.as_numpy_iterator())))
for i in list(ds_train.as_numpy_iterator())[0]:
    print(i)
    print(i.shape)
    print('*'*100)
```
